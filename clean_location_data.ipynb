```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150

filename = 'location_frame_0.txt'
```

```python
location = pd.read_pickle(filename)
print location.head()
```

```python
'''Clean the geocoding data by dismissing confidences that are rated poorly'''
# location = location[location.latlng_confidence.isin(['])
# def remove_poor_match(conf):
#     #if C or X are in the confidence then we dismiss the result
#     #see http://www.mapquestapi.com/geocoding/geocodequality.html
#     if 'C' in conf or 'X' in conf:
#         return False
#     return True
location = location[location.latlng_confidence.isin(['P1AAA','L1AAA'])]
'''Dismiss latitudes outside of Bay Area'''
correct_latitude = pd.np.logical_and(37.3333 <= location.latitude, location.latitude <= 37.9736)
correct_longitude = pd.np.logical_and(-122.5311 <= location.longitude, location.longitude <= -121.9000)
location = location[pd.np.logical_and(correct_latitude, correct_longitude)]
```

```python
print location.groupby('latlng_confidence').count()
```

```python
plt.plot(location.longitude, location.latitude, '*', alpha = 0.02)
plt.ylim(37.3333,37.9736)
plt.xlim(-122.5311, -121.9000)
```
