```python
'''Importing igraph'''
import igraph as ig
```

```python
g = ig.Graph(directed = True)
```

```python
g.add_vertices(3)
```

```python
g.add_edges([(0,1),(1,2),(2,1)])
```

```python
print g
```

```python
print g.degree()
print g.edge_betweenness()
```

```python
g.vs["name"] = ["Alice","Bob","Charlie"]
```

```python
g.vs.find("Alice").degree()
```

```python
visual_style = {"bbox":(300,300),
                "margin":50,
                "vertex_label":g.vs["name"],
                "vertex_size":25,
                "layout": g.layout("kk"),
                "vertex_label_size":15,
                'vertex_label_angle':1,
                "vertex_color": ["orange","red","green"],
                "vertex_label_dist": 2.0,
}
ig.plot(g, **visual_style)
```
