```python
'''Importing igraph'''
import igraph as ig
import pandas as pd
np = pd.np
```

```python
df = pd.read_pickle('for_visualization.pickle')
```

```python
depth = 500
starting_point = 'town-san-carlos'

g = ig.Graph(directed = True)
g.add_vertex(starting_point)

def fill_connections(starting_point):
    for to_id in df[df.from_yelp_id == starting_point].to_yelp_id:
        new = not to_id in g.vs['name']
        if new:
            g.add_vertex(to_id)
        g.add_edge(starting_point, to_id)
        if new:
            fill_connections(to_id)
        
fill_connections(starting_point)
```

```python

np.array(g.vs['name'])[np.array(g.degree(type="in")).argsort()]
```

```python
def color_picker(name):
    if 'palo-alto' in name:
        return 'blue'
    elif 'san-carlos' in name:
        return 'black'
    elif 'san-mateo' in name:
        return 'red'
    elif 'redwood-city' in name:
        return 'orange'
    else:
        return 'grey'
    
    
    
    


visual_style = {"bbox":(1920,1080),
                "margin":50,
#                 "vertex_label":g.vs["name"],
                "vertex_size":20,
                "layout": g.layout("kk"),
#                 "vertex_label_size":2,
                "vertex_color": [color_picker(n) for n in g.vs["name"]],
#                 "vertex_label_dist": 2.0,
                "edge_arrow_size":1,
                "edge_arrow_width":0.5,
}
ig.plot(g, 'graph.pdf', **visual_style)
```

```python
def find_competitors(graph):
    '''Finds pairs of vertices with most common parent nodes'''
    common_parents = []
    for g1 in g.vs:
        for g2 in g.vs:
            if g1 < g2:
                g1_parents = g.neighbors(g1['name'], mode = 'in')
                g2_parents = g.neighbors(g2['name'], mode = 'in')
                common = set(g1_parents).intersection(set(g2_parents))
                if len(common) > 0:
                    common_parents.append((g1['name'], g2['name'], len(common)))
    return common_parents

comps = find_competitors(g)
```

```python
comps = sorted(comps, key = lambda x: x[2])
```
